```python
###Librerias Necesarias:

import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import statsmodels.stats.outliers_influence as oi
import statsmodels.graphics.api as smg
import statsmodels.stats.api as sms
!pip install pandas

import pandas as pd #librería para manejo de datos
```

```python
### Carga de datos:

datos= pd.read_csv('EcommerceCustomers.csv', sep=",") #Lee parte de la base de datos
datos.columns
```

```python
### Modelo de Mínimos Cuadrados Ordinarios

mco = smf.ols("GastoAnual ~ TiempoApp + TiempoWeb + MembresíaAños ", data=datos).fit()
mco.summary()
```

```python
##Calculo de Factor Inflación de la Varianza

X= datos[["TiempoApp", "TiempoWeb", "MembresíaAños"]]
FIVS=[oi.variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print("FIVs: ", FIVS)

### Cálculo del Número de condición

NC = np.sqrt(mco.condition_number)
print("Numero de Condicion: ", NC)

#Matriz de Correlaciones

corr_matrix=np.corrcoef(X.T)
smg.plot_corr(corr_matrix, xnames=["TiempoApp", "TiempoWeb", "MembresíaAños"])
plt.show()
```

```python
#GOLDFELD-QUANDT
GQ=sms.het_goldfeldquandt(mco.model.endog, mco.model.exog, split=100)
print("GOLDFELD-QUANDT: ", GQ)

#BREUSH-PAGAN
BP=sms.het_breuschpagan(mco.resid, mco.model.exog)
print("BREUSH-PAGAN: ", BP)

#WHITE
W=sms.het_white(mco.resid, mco.model.exog)

print("White: ", W)
```
